```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/8-thsem -internship/cleaned_superstore_dataset.csv")
df.head()
```

```python
total_sales = df["sales"].sum()
total_profit = df["profit"].sum()
profit_margin = total_profit / total_sales

print(total_sales, total_profit, profit_margin)
```

```python
segment_analysis = df.groupby("segment")[["sales","profit"]].sum()
print(segment_analysis)
```

```python
category_analysis = df.groupby("category")[["sales","profit"]].sum()
print(category_analysis)
```

```python
region_analysis = df.groupby("region")[["sales","profit"]].sum()
print(region_analysis)
```

```python
discount_analysis = df.groupby("discount")["profit"].mean()
print(discount_analysis)
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])

monthly_sales = df.groupby(df["order_date"].dt.month)["sales"].sum()
print(monthly_sales)
```
