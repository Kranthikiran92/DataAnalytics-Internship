```python
# Import necessary libraries for data manipulation
import pandas as pd
import numpy as np
```

```python
# Load the dataset from the specified CSV file into a pandas DataFrame and display the first few rows
df = pd.read_csv("/content/drive/MyDrive/8-thsem -internship/SuperStoreOrders - SuperStoreOrders.csv")

df.head()
```

```python
# Display the data types of each column in the DataFrame
df.dtypes
```

```python
# Display the shape (number of rows and columns), column names, and information about the DataFrame including data types and non-null values
df.shape
df.columns
df.info()
```

```python
# Check for missing values in each column and sum them up
df.isnull().sum()
```

```python
# Check for duplicate rows in the DataFrame and sum them up
df.duplicated().sum()
```

```python
# Convert 'order_date' and 'ship_date' columns to datetime objects, specifying dayfirst=True for correct parsing
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)
df['ship_date'] = pd.to_datetime(df['ship_date'], dayfirst=True)
```

```python
# Calculate the number of delivery days by subtracting 'order_date' from 'ship_date' and extracting the days component
df["delivery_days"] = (df["ship_date"] - df["order_date"]).dt.days
```

```python
# Convert 'sales' and 'profit' columns to numeric types, coercing any errors to NaN
df["sales"] = pd.to_numeric(df["sales"], errors="coerce")
df["profit"] = pd.to_numeric(df["profit"], errors="coerce")
```

```python
# Calculate the profit margin by dividing 'profit' by 'sales'
df["profit_margin"] = df["profit"] / df["sales"]
```

```python
# Recalculate profit margin, replacing 0 sales values with NaN to avoid division by zero errors
df["profit_margin"] = df["profit"] / df["sales"].replace(0, np.nan)
```

```python
# Create a new column 'sales_category' by categorizing 'sales' into bins: Low, Medium, High, Very High
df["sales_category"] = pd.cut(df["sales"],
                             bins=[0,100,500,1000,5000],
                             labels=["Low","Medium","High","Very High"])
```

```python
# Define the path to save the cleaned dataset in Google Drive
drive_path = "/content/drive/MyDrive/8-thsem -internship/cleaned_superstore_dataset.csv"

# Save the cleaned DataFrame to the specified Google Drive path
df.to_csv(drive_path, index=False)

print(f"Cleaned dataset saved to: {drive_path}")
```
