```python
import pandas as pd
from scipy import stats

df = pd.read_csv("/content/drive/MyDrive/8-thsem -internship/cleaned_superstore_dataset.csv")
```

```python
low_discount = df[df["discount"] <= 0.2]["profit"]
high_discount = df[df["discount"] > 0.2]["profit"]
```

```python
t_stat, p_value = stats.ttest_ind(low_discount, high_discount)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
Rule:

If p-value < 0.05 → Reject H0

Example Interpretation:

The p-value is less than 0.05, which indicates that discounts have a statistically significant impact on profit. Higher discounts lead to lower profit.
```
